A data analysis of the declining birth rates in the UK and multiple developing countries, UK teen pregnancy rates and fertility rates.

Are declining birth rates going to effect these countries in anyway?

Are UK teen pregnancy rates high?

The purpose of this project is to delve in to the declining birth rates of the UK and certain developing countries such as Yeman, Rwanda, Kenya, etc starting from the year 1950 up until the year 2020 and to discover how the decline will effect these countries. Also this project will look into UK teen pregnancy rates to see if there has been a decline or increase. The project will also take a look at UK fertility rates and find which age group has the highest fertility. 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('birth rates.csv')
birth_rate=df.copy()
birth_rate.copy()
```

Dropping columns

```python
birth_rate.drop('% - change in years',axis=1,inplace=True)
birth_rate.head()
```

```python
# Sample Line Graph for Selected Countries
countries = ['UK', 'Yemen', 'Rwanda', 'Kenya', 'Afghanistan', 'Philippines', 'Iran', 'Egypt', 'Bangladesh', 'India']

# Filter data for selected countries
filtered_data = birth_rate[birth_rate['Country'].isin(countries)]

# Pivot data to make countries columns for plotting
pivot_data = filtered_data.pivot(index='Year', columns='Country', values='Number of children per woman')

# Plotting the line graph
pivot_data.plot(kind='line', figsize=(12, 8))
plt.xlabel('Year', fontsize=15)
plt.ylabel('Number of children per woman', fontsize=15)
plt.title('Birth rates of the UK and multiple developing countries over time', fontsize=18)
plt.legend(title='Country')
plt.grid(True)
plt.show()
```

As the graph shows above the developing countries have always had a high rate of children per woman compared to the UK, this is due to the poor education they receive and no birth controls or protection available. Iran which is represented by the purple line shows that they had the biggest decline of birth rates in comparison to the other developing countries until 2010 where it began to increase again. This is because the government implemented economic incentives and pro natalist policies to encourage larger families. The UK has started to decline in birth rates due to many factors, for example the cost of living, more women are choosing to have a career instead of a family, and some women wait till they are in their 40's to have a baby and then discover they have fertility problems and can't conceive. If the birth rate continues to decline worldwide, this will have a grand effect, such as labour shortages, strains on the social and healthcare system and also a shrinking workforce. The developing countries with decreased birth rates will also be better able to keep up with funding for schools, healthcare and infrastructure. In addition to this the ratio of labor force to children will go up, resulting in less children workng instead of being in education. 

An analysis of UK teen pregnancy from 1920 - 1995.

```python
df=pd.read_csv('Teen pregnancy.csv')
teen_births=df.copy()
teen_births.head()
```

Renaming columns

```python
teen_births.rename(columns=({'Year of birth of woman': 'Year', 'Percentage of women who have had a child by their 20th birthday': 'Percentage of teen pregnancy'}), inplace=True)
teen_births.head()
```

UK teen pregnancy rates 1920 - 1995.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Filter data to show only every 5 years
teen_births_filtered = teen_births[teen_births['Year'] % 5 == 0]

# Create horizontal bar graph
plt.figure(figsize=(12, 8))
sns.barplot(
    y='Year',
    x='Percentage of teen pregnancy',
    data=teen_births_filtered,
    errorbar=None,
    orient='h',
    hue='Year',  # Assign the 'Year' variable to hue
    palette='flare',  # Add the palette here
    legend=False  # Disable the legend for simplicity
)

# Set labels and title
plt.xlabel('Percentage of Teen Pregnancy', fontsize=18)
plt.ylabel('Year', fontsize=18)
plt.title('UK Teenage Pregnancy Rates by 5-Year Intervals (1920–1995)', fontsize=18)

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

As per shown in the above graph, 1950-1955 had the highest rate of teenage pregnancy during the years of 1920 - 1995. This is partly due to the post war baby boom, and also because there was not many birth controls available in that era, neither sex education was rarely taught. The year 1995 starts to decrease because there was improved contraceptive use and a multiple use of different contraceptions.

UK teen birth rates (2019-2024)

```python
df=pd.read_csv('Teen birth rates.csv')
teen_rates=df.copy()
teen_rates.copy()
```

```python
import matplotlib.pyplot as plt

# Line chart for UK teen birth rates
plt.figure(figsize=(8, 5))
plt.plot(teen_rates['Year'], teen_rates['Number of births per 1000 teens'], marker='o', color='mediumvioletred')

plt.title('UK Teen Birth Rates (2019–2023)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Births per 1000 Teens', fontsize=12)
plt.grid(True)
plt.xticks(teen_rates['Year'])  # Ensures all years are labeled
plt.tight_layout()

plt.show()
```

As seen in the graph above UK teen pregnancy spiked during 2022 because during the covid lock downs in 2020 and 2021, access to sexual health services and education was very limited which impacted the teen pregnancy rates. Also areas with the highest deprivation seen that they had the highest levels of pregnant teenagers in comparison to the least deprived areas.

Quantity of births divided by 5 year age gap.

```python
df=pd.read_csv('Number of births by 5 year age gap.csv')
age_groups=df.copy()
age_groups.head()
```

Visualizing the data and extracting columns (1950 - 2020)

```python
import matplotlib.pyplot as plt

# Filter data for years divisible by 10
filtered_data = age_groups[age_groups['Year'] % 10 == 0]

# Extract relevant columns
age_groups = ["Age 10 - 14", "Age 15 - 19", "Age 20 - 24", "Age 25 - 29"]
filtered_data = filtered_data[['Year'] + age_groups]

# Plot the bar chart
plt.figure(figsize=(10, 6))
x = filtered_data['Year']
width = 2.5  # Bar width

# Create bars for each age group
for i, age_group in enumerate(age_groups):
    plt.bar(x + i * width, filtered_data[age_group], width, label=age_group)

# Formatting the plot
plt.xlabel('Year')
plt.ylabel('Number of Births')
plt.title('Number of Births by Age Group (Years Divisible by 20)')
plt.xticks(x + width * 1.5, x, rotation=45)
plt.legend(title="Age Groups")
plt.tight_layout()

# Show the plot
plt.show()
```

As you can see from the bar graph the most average age group to have a child was between the ages of 20 - 24, apart from the year 2020. The age gap 20 -24 was very popular because fertility is high during these ages and there is also fewer complications. The age of women having children started to increase in 2020 due to the Covid-19 pandemic, and also because some women wanted to experience life more such as travelling and making a career to be financially stable. Also alot of women have anxiety about economic activities and job stability.

Fertility rates of women grouped by age and how many births there was in each age group out of every 1000 births in England and Wales (1935 - 1994).

```python
df=pd.read_csv('Fertility rates.csv')
fertility_rates=df.copy()
fertility_rates.head()
```

```python
import matplotlib.pyplot as plt

fertility_rates.columns = fertility_rates.columns.str.strip()

age_columns = ['Age 20', 'Age 25', 'Age 30', 'Age 35', 'Age 40']

average_births = fertility_rates[age_columns].mean()

plt.figure(figsize=(5, 5))
plt.pie(average_births, labels=age_columns, autopct='%1.2f%%', startangle=140)
plt.title('Average Fertility Rate by Age Group (1935 - 1994)')
plt.axis('equal') 
plt.show()
```

As the pie chart shows above the average age of fertility during these years was the age 25. This age group was very popular due to a combination of social and economic factors. Women of 25 years old were ready to start families by this age, as they had just finished education or establishing a career, and were still young enough to have multiple children. Societal norms also had a big influence on women encouraging them to marry and have children within a time frame that aligned with the age of mid 20's. Also science say's that women are most fertile in their 20s and 25 is the peak of that fertility window.
